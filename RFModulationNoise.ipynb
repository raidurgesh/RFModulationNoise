```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 100)
RFModNoise=(6e-5*(1+(7e-2/freq)))**2
plt.style.use('bmh')
plt.plot(freq, RFModNoise, 'k-')
plt.yscale("log")
plt.xscale("log")
plt.title("RF Modulation Noise")
plt.xlabel("Frequency (Hz)")
plt.ylabel("PSD (rad/Hz)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 100)
RFModNoise=(6e-5*(1+(7e-2/freq)))**1
plt.style.use('bmh')
plt.plot(freq, RFModNoise, 'k-')
plt.yscale("log")
plt.xscale("log")
plt.title("RF Modulation Noise")
plt.xlabel("Frequency (Hz)")
plt.ylabel("ASD (rad/sqrt(Hz)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 100)
RFModNoise_PSD=(6e-5*(1+(7e-2/freq)))**2
RFModNoise_ASD=(6e-5*(1+(7e-2/freq)))
plt.style.use('bmh')
plt.plot(freq, RFModNoise_ASD,'b-',freq, RFModNoise_PSD,'r--')
plt.yscale("log")
plt.xscale("log")
plt.title("RF Modulation Noise")
plt.xlabel("Frequency (Hz)")
plt.ylabel("PSD(rad/Hz) | PSD(rad/Hz)")
plt.show()
```
